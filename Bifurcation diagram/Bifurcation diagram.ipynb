```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def LastXs(x0=0.2, r=2.0, n=1000, last=100):
    x = [x0]
    for i in range(n-1):
        x += [r * x[i] * (1 - x[i])]
    return x[-last:]
```

```python
xs = LastXs(r=3.05, n=1000, last=1000)
plt.figure(figsize=(20,5))
plt.title('Evolution of x, r=3.05')
plt.xlabel('Time')
plt.ylabel('x_n')
plt.plot(np.arange(1,1001), xs)
```

```python
n = 10000
rs = []
xs = []
last = 1000

for r in np.linspace(0, 4, 10000):
    xs += LastXs(r=r, n=n, last=last)
    rs += [r]*last
```

```python
plt.figure(figsize=(20, 10))
plt.title('Bifurcation diagram')
sns.scatterplot(rs, xs, s=10, linewidth=0.07)
```
